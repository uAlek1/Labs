```python
#cложение точек
#input: first_point: list, where [0] = x-coordinate, [1] = y-coordinate
#       second_point: list, where[0] = x-coordinate, [1] = y-coordinate
#       elliptic: polynom, defining an elliptic curve in a factor ring

#output: true - if unless an exception has occurred And point
#        false -  otherwise And denominator of the angular coefficient

def add(first_point:list, second_point:list, elliptic)->list:
    if first_point[0] != second_point[0]:
        #print("(")
        denominator = second_point[0] - first_point[0]
        #print(f"{second_point[0] = }")
        #print(f"{first_point[0] = }")
        #print(f"{denominator = }")
        result = []
        try:
            inv_den = 1/denominator
            #print(f"{1/denominator = }")
            #print(type(second_point[1]))
           # print(type(first_point[1]))
            obliquity = (second_point[1] - first_point[1]) * inv_den
            result.append(obliquity**2 * elliptic  - first_point[0] - second_point[0])# 
            result.append(obliquity*(first_point[0] - result[0]) - first_point[1])
            return True, result
        except ZeroDivisionError:
         #   print("except")
            return False, denominator
        
    elif (first_point[1] == (-second_point[1])):
        print("0-point")
        return True, [0,0]
    
    else:
      #  print(")")
        denominator = 2 * first_point[1] * elliptic
        result = []
        try:
            inv_den = 1/denominator
            #print(f"{1/denominator = }")
            obliquity = (3*first_point[0]^2 + A) * inv_den
            result.append(obliquity^2 * elliptic - 2 * first_point[0])
            result.append(obliquity * (first_point[0]-result[0]) - first_point[1])
            return True, result
        except ZeroDivisionError:
            return False, denominator
```

```python
def get_division_poly(k: int, E:EllipticCurve):
    fk = E.division_polynomial(k,two_torsion_multiplicity=0)
    fkp1 = E.division_polynomial(k+1,two_torsion_multiplicity=0)
    fkp2 = E.division_polynomial(k+2,two_torsion_multiplicity=0)
    if(k-1 == 0):
        fkm1 = 0
    else: 
        fkm1 = E.division_polynomial(k-1,two_torsion_multiplicity=0)
    if(k-1 == -1):
        fkm1 = -1
    if(k-2 == 0):
        fkm2 = 0
    else:
        fkm2 = E.division_polynomial(k-2,two_torsion_multiplicity=0)
    if(k-2 == -1):
        fkm2 = -1
    #print(f"{fk = }, {fkp1 = }, {fkp2 = }, {fkm1 = }, {fkm2 = }")
    return fk,fkp1,fkp2,fkm1,fkm2
```

```python
def get_S(E, k, poly):
    R.<x> = PolynomialRing(GF(p))
    FR.<a> = R.quotient(poly)
    elliptic = a^3 + A*a + B
    first_point = []
    first_point.append(a**(p**2))
    first_point.append(elliptic**((p**2-1)//2))
    second_point = []
    fk,fkp1,fkp2,fkm1,fkm2 = get_division_poly(k, E)
    if(k&1):
        second_point.append(FR(((a * fk**2 - 4 * elliptic * fkm1 * fkp1)/ (fk**2) )))
        second_point.append(FR(((fkp2*fkm1**2 - fkm2*fkp1**2 +(a-a))/ (fk**3)) )) 
      #  print(f"123123{second_point=} ")
    else:
        second_point.append(FR(((4*a*fk**2 * elliptic - fkm1*fkp1)/(4*fk**2*elliptic))))
        second_point.append(FR((fkp2*fkm1**2 - fkm2*fkp1**2)/(16*fk**3*elliptic**2)))
   #     print(f"123123123{second_point = }")
  #  print(f"{first_point = }")
   # print(f"{second_point = }")
    success, result_point = add(first_point, second_point, elliptic)
    if success:
        return result_point, poly
    else:
        #print(f"{result_point = }")
        gcd = poly.gcd(result_point.lift())
        #print(f"{poly = }")
        #print(f"{gcd = }")
        return get_S(E, k, gcd)
    
def get_F(E, k, poly):
    R.<x> = PolynomialRing(GF(p))
    FR.<a> = R.quotient(poly)
    elliptic = a^3 + A*a + B
    result = []
    result.append(a**p)
    result.append(elliptic**((p-1)//2))
    return result  

def get_Ft(E,k,poly, F):
    R.<x> = PolynomialRing(GF(p))
    FR.<a> = R.quotient(poly)
    elliptic = a^3 + A*a + B
    result = []
    fk,fkp1,fkp2,fkm1,fkm2 = get_division_poly(k, E)
    if k&1:
        result.append(FR((F[0] * fk**(2*p) - 2**(2*p) *elliptic**p *fkm1**p *fkp1**p) / fk**(2*p)))
        result.append(FR((2**(2*p-2) *elliptic**(p-1) *(fkp2**p *fkm1**(2*p) -fkm2**p *fkp1**(2*p))) / (F[1] * fk**(3*p))))
        return result
    else:
        result.append(FR((4**p * F[0] * fk**(2*p) * elliptic**p - fkm1**p * fkp1**p) / (4**p * fk**(2*p) * elliptic**p)))
        result.append(FR((fkp2**p * fkm1**(2*p) - fkm2**p * fkp1**(2*p)) / (2**(2*p+2) * F[1] * fk**(3*p) * elliptic**(p+1))))
        return result
```

```python
from functools import reduce
from operator import mul  

def chinese_remainder(n, a): 
    s = 0
    prod = reduce(mul, n)
    for n_i, a_i in zip(n, a):
        p = prod // n_i 
        s += a_i * inverse(p, n_i) * p 
    return s % prod 

def inverse(a, b):
    b0 = b
    x0, x1 = 0, 1
    if b == 1:
        return 1
    while a > 1:
        q = a // b
        a, b = b, a % b
        x0, x1 = x1 - q * x0, x0
    if x1 < 0:
        x1 += b0
    return x1 
```
