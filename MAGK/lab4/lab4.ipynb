```python
#трехсторонний протокол установоения ключа
#Есть три пользователя A,B,C, им известна кривая и точка P
#Они выбирают каждый случайное число A-a, B-b, C-c, и считают PA = a*P, PB = b*P, PC = c*P и отправляют эти значения
#                    A
#             bP -     - cP
#        - -              -  -
#   - aP                       aP  -
#B    cP   =   =   =   =   =   bP   C

#Необходимо выбрать такую функцию F, чтобы F(a,PB,PC) = F(b,PA,PC) = F(c, PA,PB) и ее значение трудно было вычислить, зная PA,PC,PB
#F(x,P,Q) = e(P,Q)^x
#тогда F(a,PB,PC) = F(a,PB,PC) = F(a,PB,PC) = F(1,P,P)^abc
# F(a,PB,PC) = e(PB,PC)^a = e(bP, cP)^a = e(P,P)^abc
# F(b,PA,PC) = e(PA,PC)^b = e(aP, cP)^b = e(P,P)^abc
# F(c,PA,PB) = e(PA,PB)^c = e(aP, bP)^c = e(P,P)^abc
# F(1,P,P)^abc = e(P,P)^abc
```

```python
import random

class Proto:
    def __init__(self, P, Q, E, p):
        self.P = P
        self.Q = Q
        self.E = E
        self.p = p
        self.recvPoints = []
        self.randomNum = 0
    
    def __generatePoint(self):
        self.randomNum = random.randint(1, self.p - 1)
        pointP = self.randomNum * self.P
        pointQ = self.randomNum * self.Q
        return pointP, pointQ
    
    def send(self):
        return self.__generatePoint()
    
    def recv(self, pointP, pointQ):
        self.recvPoints.append([pointP, pointQ])
    
    def __key(self):
        if(len(self.recvPoints)) != 2:
            return 0
        key = self.E(self.recvPoints[0][0]).weil_pairing(self.E(self.recvPoints[1][1]), r)
        key = key ** self.randomNum
        return key
    
    def getKey(self):
        return self.__key()
    
```

```python
A = Proto(P, Q, EE, r)
B = Proto(P, Q, EE, r)
C = Proto(P, Q, EE, r)

PA,QA = A.send()
PB,QB = B.send()
PC,QC = C.send()

A.recv(PB, QB)
A.recv(PC, QC)
B.recv(PA, QA)
B.recv(PC, QC)
C.recv(PA, QA)
C.recv(PB, QB)

print(f"{A.getKey() = }")
print(f"{B.getKey() = }")
print(f"{C.getKey() = }")
```

```python
class BN(object):
    @staticmethod
    def generate_prime_order(zbits):
        while True:
            z = randint(2^(zbits - 1), 2^zbits)
            pz = int(BN.p(z))
            if not is_prime(pz):
                continue
            rz = int(BN.r(z))
            if not is_prime(rz):
                continue
            break
        K = GF(pz)
        b = 1
        while True:
            curve = EllipticCurve(K, [0, b])
            card = curve.cardinality()
            if card % rz == 0:
                break
            b += 1
        return curve, pz, b

    @staticmethod
    def p(z):
        return 36 * z^4 + 36 * z^3 + 24 * z^2 + 6 * z + 1
    @staticmethod
    def r(z):
        return 36 * z^4 + 36 * z^3 + 18 * z^2 + 6 * z + 1
    @staticmethod
    def t(z):
        return 6 * z^2 + 1
```
